## Analysis of Bigrams

1. Take any corpus(article) of your choice and analyze all the bigrams present based on ConditionalFrequencyDist and compute the co-occurence matrix.
2. Visualize the co-occurence matrix.

```python
import nltk
import numpy as np
import pandas as pd
import plotly.offline as py
import plotly.graph_objs as go
import nltk
py.offline.init_notebook_mode(connected=True)
```

```python
text="""I love iphone.  
I am so in love with iphone. 
iphone is great. 
samsung is great. 
iphone is bad. 
I really really love iphone cases. 
samsung can never beat iphone. 
samsung is better than apple."""
```

# BIGRAMS #

```python
bgram=list(nltk.bigrams(nltk.tokenize.word_tokenize(text)))
```

```python
bgram
```

```python
bgram_cfd = nltk.ConditionalFreqDist(nltk.bigrams(nltk.tokenize.word_tokenize(text)))
```

```python
bgram_cfd
```

```python
bgram_cfd_df=pd.DataFrame(bgram_cfd)
bgram_cfd_df.fillna(0,axis=1,inplace=True)
```

```python
bgram_cfd_df
```

```python
co_occ_matrix=bgram_cfd_df.as_matrix()
```

```python
co_occ_matrix
```

```python
col=list(bgram_cfd_df.columns)
trace = go.Heatmap(z=co_occ_matrix.tolist(),
                   x=col,
                   y=list(bgram_cfd_df.index))
data=[trace]
py.iplot(data, filename='labelled-heatmap')
```

# Trigrams #

```python
trigram=list(nltk.trigrams(nltk.tokenize.word_tokenize(text)))
```

```python
trigram
```

```python
trigram_cfd = nltk.ConditionalFreqDist([('('+ w1 + ' ' + w2+ ')', w3) for w1, w2, w3 in nltk.trigrams(nltk.word_tokenize(text.lower()))])
```

```python
trigram_cfd
```

```python
trigram_cfd_df = pd.DataFrame(trigram_cfd).transpose()
trigram_cfd_df.fillna(0,axis=1,inplace=True)
```

```python
trigram_cfd_df
```

```python
co_occ_matrix=trigram_cfd_df.as_matrix()
```

```python
co_occ_matrix
```

```python
col=list(trigram_cfd_df.columns)
trace = go.Heatmap(z=co_occ_matrix.tolist(),
                   x=col,
                   y=list(trigram_cfd_df.index))
data=[trace]
py.iplot(data, filename='labelled-heatmap')
```

# Tags #

```python
nltk.tag.pos_tag(nltk.tokenize.word_tokenize("I am in a good mood."))
```

```python
tag_sent=""
for sent in nltk.tokenize.sent_tokenize(text):
    for word_tags in nltk.tag.pos_tag(nltk.tokenize.word_tokenize(sent)):
        tag_sent=tag_sent+word_tags[1]+" "        
```

```python
tag_bgram=list(nltk.bigrams(nltk.tokenize.word_tokenize(tag_sent)))
```

```python
tag_bgram
```

```python
tag_bgram_cfd = nltk.ConditionalFreqDist(nltk.bigrams(nltk.tokenize.word_tokenize(tag_sent)))
```

```python
tag_bgram_cfd
```

```python
tag_bgram_cfd_df=pd.DataFrame(tag_bgram_cfd)
tag_bgram_cfd_df.fillna(0,axis=1,inplace=True)
```

```python
tag_bgram_cfd_df
```

```python
co_occ_matrix=tag_bgram_cfd_df.as_matrix()
```

```python
co_occ_matrix
```

```python
col=list(tag_bgram_cfd_df.columns)
trace = go.Heatmap(z=co_occ_matrix.tolist(),
                   x=col,
                   y=list(tag_bgram_cfd_df.index))
data=[trace]
py.iplot(data, filename='labelled-heatmap')
```
