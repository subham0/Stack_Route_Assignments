## Level 2 - Word Prediction using LSTM

Follow the same steps as in Char Prediction (Level 1) but at the word level than at the Char Level. 

```python
from keras.layers import LSTM, Dense
from keras.layers import Embedding
from keras.models import Sequential
from keras.utils import np_utils
import numpy as np
import sklearn.model_selection as m_sel
import nltk
from nltk.tokenize import sent_tokenize,word_tokenize
import re
```

```python
string = open('345-8.txt').read()
new_str = re.sub('[^a-zA-Z]',' ', string)
open('clean_data.txt', 'w').write(new_str)
words = []
with open("clean_data.txt", 'r') as in_file:
    text = in_file.read()
    sents = nltk.sent_tokenize(text)
    for i in range(0,len(sents),1):
        for j in word_tokenize(sents[i]):
            words.append(j)
```

```python
unique, counts = np.unique(words, return_counts=True)
words_freq_dic = dict((unique[i],counts[i]) for i in range(0,len(unique),1))
```

```python
def tell_freq(x,words_freq_dic):
    if x in words_freq_dic.keys():
        return words_freq_dic[x]
```

```python
word_freq = [tell_freq(word,words_freq_dic) for word in words]    
```

```python
min(word_freq)
```

```python
max(word_freq)
```

```python
seq_in = []
seq_out = []
c = 1
inp = []
for freq in word_freq:
    if c <= 100:
        inp.append(freq)
        c = c + 1
    elif c == 101:
        seq_in.append(inp)
        inp = []
        seq_out.append(freq)
        c = 1        
```

```python
np.shape(seq_in)
```

```python
np.shape(seq_out)
```

```python
x = np.array(seq_in)
```

```python
y = np.array(seq_out)
```

```python
np.shape(x)
```

```python
np.min(x)
```

```python
np.max(y)
```

```python
np.min(y)
```

```python
num_alpha = np.max(y)
```

```python
ycat = np_utils.to_categorical(y, num_classes=num_alpha+1)
```

```python
model = Sequential()
model.add(Embedding(num_alpha+1, 256, input_length = 100))
model.add(LSTM(128,return_sequences=True))
model.add(LSTM(128))
model.add(Dense(num_alpha+1, activation = "sigmoid"))
model.compile(loss = "categorical_crossentropy", optimizer = "adam", metrics = ["accuracy"])
model.fit(x, ycat, batch_size = 25, epochs = 10)
```
